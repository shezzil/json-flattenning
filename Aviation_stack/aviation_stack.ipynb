```python

import requests

url = "https://api.aviationstack.com/v1/flights?access_key=a6c3914b3cefad756352fcebb176ed26"

response = requests.get(url)

data = response.json()
data
```

```python
flights = data['data']
flights
```

```python
# Flattening for Flight Table
import pandas as pd

flights_data = []

for idx, flight in enumerate(flights, start=1):
    row = {
        'flight_number': flight.get('flight', {}).get('number'),
        'flight_date': flight.get('flight_date'),
        'flight_status': flight.get('flight_status'),
        'departure_iata': flight.get('departure', {}).get('iata'),
        'arrival_iata': flight.get('arrival', {}).get('iata'),
        'airline_iata': flight.get('airline', {}).get('iata'),
        'flight-iata' : flight.get('flight', {}).get('iata'),
        'delay_minute': flight.get('arrival', {}).get('delay')
    }
    flights_data.append(row)

df_flights_data = pd.DataFrame(flights_data)
df_flights_data.head(26)
```

```python
# Flattening for airport table

airport_data = []

for idx, airport in enumerate(flights, start=1):

    row_2 = {
        'airport_name' : airport.get('departure').get('airport'),
        'iata' : airport.get('departure').get('iata'),
        'icao' : airport.get('departure').get('icao'),
        'time_zone' : airport.get('departure').get('timezone'),
        
    }

    airport_data.append(row_2)

df_airport_data = pd.DataFrame(airport_data).drop_duplicates()
df_airport_data
```

```python
# Flattening for time zone

timezone_data = []

for timezone in flights:
    row_3 = {
        'flight_iata' : timezone.get('flight', {}).get('iata'),
        'dep_scheduled' : timezone.get('departure').get('scheduled'),
        'dep_estimated' : timezone.get('departure').get('estimated'),
        'dep_actual' : timezone.get('departure').get('actual'),
        'arr_scheduled' : timezone.get('arrival').get('scheduled'),
        'arr_estimated' : timezone.get('arrival').get('estimated'),
        'arr_actual' : timezone.get('arrival').get('actual')
    }

    timezone_data.append(row_3)

df_timezone_data = pd.DataFrame(timezone_data)
df_timezone_data.head(26)
```

```python
# Flatenning airline table

airline_data = []

for airline in flights:
    row_4 = {
        'airline_name' : airline.get('airline').get('name'),
        'iata' : airline.get('airline').get('iata'),
        'icao' : airline.get('airline').get('icao'),
    }
    
    airline_data.append(row_4)

df_airline_data = pd.DataFrame(airline_data).drop_duplicates()
df_airline_data
```

```python
import pandas as pd
import json
```

```python
data = pd.read_json("C:/My Files/Courses/Python/NamasteSQL_Class_Notes/Day_8/assignment_orders.json")
data
```

```python
f = open("C:/My Files/Courses/Python/NamasteSQL_Class_Notes/Day_8/assignment_orders.json","r")
data = json.load(f)
data
```

```python
order_data = []

for order in data:
    for order_loop in order['orders']:
        for product_loop in order_loop['products']:
            row = {
                'order_id' : order_loop['order_id'],
                'order_date' : order_loop['order_date'],
                'total_amount' : order_loop['total_amount'],
                'customer_id' : order['customer_id'],
                'product_id' : product_loop['product_id'],
                'quantity' : product_loop['quantity']
            } 
            
            order_data.append(row)
df_orders = pd.DataFrame(order_data)
df_orders
```

```python
product_data = []

for order in data:
    for order_loop in order['orders']:
        for product_loop in order_loop['products']:
            row = {
                'order_id' : order_loop['order_id'],
                'product_id' : product_loop['product_id'],
                'name' : product_loop['name'],
                'category': product_loop['category'],
                'price' : product_loop['price']
            }

            product_data.append(row)

df_product = pd.DataFrame(product_data) 
df_sorted = df_product.sort_values(['product_id', 'order_id'])
df_ProductDim = df_sorted.drop_duplicates('product_id', keep = 'last') # safe to keep = 'Last', because some item might have been updated
df_ProductDim.drop(columns='order_id',inplace=True)
df_ProductDim
```

```python
customer_data = []

for customer in data:
    row = {
        'customer_id' : customer['customer_id'],
        'cust_name' : customer['name'],
        'email_id' : customer['email'],
        'address' : customer['address']
    }
    customer_data.append(row)
    
    pd.DataFrame(customer_data)
    
```
